```python
!pip install google-colab googletrans==4.0.0-rc1 gtts

from google.colab import output
from googletrans import Translator
from gtts import gTTS
import os
import speech_recognition as sr

# Record audio using Colab's built-in recording tool
def record_audio():
    print("Recording... Speak now")
    audio_path = "/content/audio.wav"
    output.record_audio(audio_path)
    print("Recording complete.")
    return audio_path

# Recognize speech from recorded audio
def recognize_speech(audio_file):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio = recognizer.record(source)
    try:
        return recognizer.recognize_google(audio)
    except sr.UnknownValueError:
        print("Speech Recognition could not understand audio.")
        return None
    except sr.RequestError as e:
        print(f"Speech Recognition error: {e}")
        return None

# Map user input language to codes
language_map = {
    'Hindi': 'hi',
    'Telugu': 'te',
    'Kannada': 'kn',
    'Tamil': 'ta',
    'Malayalam': 'ml',
    'Bengali': 'bn'
}

target_language_input = input("Enter the target language (e.g., 'Hindi', 'Telugu', 'Tamil'): ").strip()
if target_language_input not in language_map:
    print(f"Invalid language: {target_language_input}. Exiting.")
else:
    target_language_code = language_map[target_language_input]
    audio_file = record_audio()
    recognized_text = recognize_speech(audio_file)

    if recognized_text:
        print(f"Recognized Text: {recognized_text}")

        translator = Translator()
        detected_language = translator.detect(recognized_text).lang
        print(f"Detected Language: {detected_language}")

        translated_text = translator.translate(recognized_text, src=detected_language, dest=target_language_code).text
        print(f"Translated Text in {target_language_input}: {translated_text}")

        speak = gTTS(text=translated_text, lang=target_language_code, slow=False)
        audio_output = "/content/translated_audio.mp3"
        speak.save(audio_output)
        output.eval_js(f'new Audio("{audio_output}").play()')
    else:
        print("No speech input detected.")
```
